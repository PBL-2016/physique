```python
import numpy as np
import pandas as pd
import cv2
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
body_cascade = cv2.CascadeClassifier('haarcascade_mcs_upperbody.xml')
csv = pd.read_csv('addition1.csv', delimiter='\t')
```

```python
def face_width(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 3)
    if len(faces) <= 0:
        return 0
    return faces[0][2]
```

```python
def img_width(img):
    height, width, channels = img.shape[:3]
    return width
```

```python
l = []
for p in csv['画像']:
    org = cv2.imread("dataset/join/"+p, cv2.IMREAD_COLOR)
    ubd = cv2.imread("dataset/trim/join/"+p, cv2.IMREAD_COLOR)
    if ubd is not None:
        iw = img_width(ubd)
        fw = face_width(org)
        if fw > 0:
            csv.loc[(csv['画像']==p),'幅比'] = iw/fw
```
